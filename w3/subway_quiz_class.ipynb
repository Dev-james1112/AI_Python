### level0: 파일 불러와서 '관광지명'으로 인덱스 설정하기

```python
import pandas as pd

df = pd.read_csv('subway_info.csv', index_col='관광지명')
df
```

### level1: 데이터 파악, 오름차순 정리

```python
# 데이터프레임 아래에서 세 번째에 있는 관광지명은? : 강서체육공원

df.tail(3)
# 데이터프레임에 총 몇 개의 관광지가 있나요? : 601
df.index.unique()

# 1호선 노량진역에는 어떤 관광지가 있나요?
## daframe boolean 연산: and(&), or(|)
df.loc[(df['호선명'] == '1호선') & (df['역이름'] == '노량진')]

# 한강시민공원은 어떤 역에 있나요? : 암사역, 망원역
df.loc['한강시민공원', :]

# '호선명'을 기준으로 해서 오름차순으로 데이터 정렬하기
df.sort_values(by='호선명', ascending=True)
```
