# classifier algorithm -SVM

### 1. 데이터 수집 및 전처리

```python
from sklearn import datasets, svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns

iris = datasets.load_iris()
# print(dir(iris))

data = iris['data']   # 붓꽃 하나하나에 대한 정보: 꽃받침의 길이, 너비, 꽃잎의 길이, 너비
label = iris['target']    # data에 매칭되는 분류 레이블(정답이 되는 붓꽃 종)
feature_names = iris['feature_names']    # data가 의미하는 내용
target_names = iris['target_names']    # 나올 수 있는 레이블의 이름

train_data, test_data, train_label, test_label = train_test_split(data, label, random_state=0)
```

### 2. 머신러닝 모델 만들기, 학습하기 예측하기

```python
clf = svm.SVC()
clf.fit(train_data, train_label)    # 훈련 데이터셋을 사용하여 기계 학습
pre = clf.predict(test_data)    # 학습된 모델을 통해 테스트 데이터를 예측
print(pre)    # 기계 예측값
print(test_label)    # 정답

print(target_names[pre])

score = accuracy_score(test_label, pre)
print('정확도 점수: %0.2f' %score)
```
